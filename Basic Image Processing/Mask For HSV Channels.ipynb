# HSV Color Space In An Image

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
import numpy as np

%matplotlib inline
```

```python
image = cv2.imread('./images/image-2.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

### Plot color channels

```python
red = image[:,:,0]
green = image[:,:,1]
blue = image[:,:,2]
```

```python
f,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))

ax1.set_title("Red Channel")
ax1.imshow(red)

ax2.set_title("Green channel")
ax2.imshow(green)

ax3.set_title("Blue channel")
ax3.imshow(blue)
plt.show()
```

```python
plt.rcParams['figure.figsize']=[24.0,6.0]

hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_image)
plt.show()
```

```python
hue = hsv_image[:,:,0]
saturation = hsv_image[:,:,1]
value = hsv_image[:,:,2]

f,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))

ax1.set_title("Hue channel")
ax1.imshow(hue)

ax2.set_title("Saturation channel")
ax2.imshow(saturation)

ax3.set_title("Value channel")
ax3.imshow(value)

plt.show()
```

### Mask the image 

```python
lower_hue = np.array([0,55,0])
upper_hue = np.array([255,200,255])

mask_hsv = cv2.inRange(hsv_image,lower_hue,upper_hue)
plt.title("Image mask that allows Saturation in the Image")
plt.imshow(mask_hsv)
plt.show()
```

### Extract Face From Image

```python
masked_hsv_image = np.copy(image)
masked_hsv_image[mask_hsv==0]=[0,0,0]
plt.title("Face Extraction With The Help Of Saturation")
plt.imshow(masked_hsv_image,cmap='gray')
plt.show()
```
