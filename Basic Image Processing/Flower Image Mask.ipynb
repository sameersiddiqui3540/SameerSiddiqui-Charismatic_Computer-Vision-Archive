```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('../yellow_flower.png')

plt.imshow(image)

plt.show()
```

```python
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(rgb_image)

plt.show()
```

```python
image_copy = np.copy(rgb_image)

lower = np.array([0,0,0])
upper = np.array([255,255,30])

mask = cv2.inRange(image_copy,lower,upper)

plt.imshow(mask,cmap='gray')
plt.show()
```

```python
print(image.shape)
image_copy[mask == 0] = 0
plt.imshow(image_copy)
plt.show()
```

```python
blue_sky = cv2.imread('../blue_sky.jpeg')

blue_sky = cv2.cvtColor(blue_sky,cv2.COLOR_BGR2RGB)

print("original shape:",blue_sky.shape)

plt.imshow(blue_sky)

plt.show()
```

```python
blue_sky = blue_sky[0:240,0:320,:]

print("New shape:",blue_sky.shape)

plt.imshow(blue_sky)

plt.show()
```

```python
new_blue_img = blue_sky

new_blue_img[mask!=0]=0

plt.imshow(new_blue_img)

plt.show()
```

```python
complete_img = new_blue_img + image_copy

plt.imshow(complete_img)

plt.show()
```
