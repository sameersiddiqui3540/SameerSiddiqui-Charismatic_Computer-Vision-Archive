# K-means Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

%matplotlib inline

image = cv2.imread("../../practice/images/image-8.jpg")

image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```

```python
pixel_vals = image.reshape((-1,3))

pixel_vals = np.float32(pixel_vals)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,500,0.2)

k =3

retval, labels, centers = cv2.kmeans(pixel_vals,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)

centers = np.uint8(centers)

segmented_data = centers[labels.flatten()]

segmented_image = segmented_data.reshape((image.shape))

labels_reshape = labels.reshape(image.shape[0],image.shape[1])

plt.imshow(segmented_image)
plt.show()
```

```python
plt.imshow(labels_reshape==0,cmap='gray')
plt.show()
```

```python
cluster_first = 0
masked_image = np.copy(image)
masked_image[labels_reshape == cluster_first] = [0,0,255]

f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(12,6))
ax1.set_title("First Cluster")
ax1.imshow(masked_image)


cluster_second = 1
masked_image = np.copy(image)
masked_image[labels_reshape == cluster_second] = [0,255,0]

ax2.set_title("Second Cluster")
ax2.imshow(masked_image)

cluster_third = 2
masked_image = np.copy(image)
masked_image[labels_reshape == cluster_third] = [255,0,0]
ax3.set_title("Third Cluster")
ax3.imshow(masked_image)

plt.show()
```
