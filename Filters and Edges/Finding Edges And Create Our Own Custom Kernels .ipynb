# Creating A Filter For Edge Detection

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

%matplotlib inline

plt.rcParams['figure.figsize']=[24.0,6.0]

image = mpimg.imread('../../practice/images/query3.jpg')
plt.imshow(image)
plt.show()
```

### Convert the image to grayscale

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

plt.imshow(gray_image,cmap='gray')
plt.show()
```

## Aplpy Sobel Filter

```python
# 3X3 Sobel_x Filter
sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])
# 3X3 Sobel_y Filter
sobel_y = np.array([[-1,-2,-1],
                    [0,0,0],
                    [1,2,1]])

filtered_image_sobel_y = cv2.filter2D(gray_image,-1,sobel_y)

filtered_image_sobel_x = cv2.filter2D(gray_image,-1,sobel_x)

f, (ax1,ax2) = plt.subplots(1,2,figsize=(12,6))

ax1.set_title("Image After Sobel_x Filter")
ax1.imshow(filtered_image_sobel_x)

ax2.set_title("Image After Sobel_y Filter")
ax2.imshow(filtered_image_sobel_y)


plt.show()
```

## Custom 5X5 Filter

```python
plt.rcParams['figure.figsize']=[24.0,6.0]

# 5X5 sobel_y Filter
custom_kernel = np.array([[-2,-1,0,1,2],
                         [-2,-1,0,1,2],
                         [-2,-1,0,1,2],
                         [-2,-1,0,1,2],
                         [-2,-1,0,1,2]])
test_image = np.copy(image)
test_gray_image = cv2.cvtColor(test_image,cv2.COLOR_RGB2GRAY)
filtered_image = cv2.filter2D(test_gray_image,-1,custom_kernel)
plt.imshow(filtered_image)
plt.show()
```
