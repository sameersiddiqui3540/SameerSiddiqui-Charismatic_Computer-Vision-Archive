# Gaussian Blur On Image

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

%matplotlib inline

plt.rcParams['figure.figsize']=[24.0,10.0]

image = cv2.imread('../../practice/images/query4.jpg')
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
plt.show()
```

```python
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)

gray_blur = cv2.GaussianBlur(gray,(9,9),0)

f,(ax1,ax2) = plt.subplots(1,2,figsize=(20,10))

ax1.set_title("original image")
ax1.imshow(gray,cmap='gray')
ax2.set_title("Blur Image")
ax2.imshow(gray_blur,cmap='gray')
plt.show()
```

```python
sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])
sobel_y = np.array([[-1,-2,-1],
                    [0,0,0],
                    [1,2,1]])

filtered_gray = cv2.filter2D(gray,-1,sobel_x)
filtered_gray_blur = cv2.filter2D(gray_blur,-1,sobel_y)

f,(ax1,ax2) = plt.subplots(1,2,figsize=(20,10))

ax1.set_title("original gray")
ax1.imshow(filtered_gray,cmap='gray')

ax2.set_title("blur gray")
ax2.imshow(filtered_gray_blur,cmap='gray')

plt.show()
```

```python
plt.rcParams['figure.figsize']=[24.0,10.0]
retval, binary_image = cv2.threshold(filtered_gray_blur,50,255,cv2.THRESH_BINARY)

plt.title("Binary Image After Gaussian Blur")
plt.imshow(binary_image,cmap='gray')
plt.show()
```
