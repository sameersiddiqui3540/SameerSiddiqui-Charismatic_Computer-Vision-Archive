## Face detection using OpenCV

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

%matplotlib inline
```

```python
image= cv2.imread("../../practice/images/image-4.jpg")

image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(image)
plt.show()
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(20,10))
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
face_cascade_eye = cv2.CascadeClassifier('../../practice/detector_architectures/haarcascade_eye.xml')

eyes_in_faces = face_cascade_eye.detectMultiScale(gray,3.5,10)
```

```python
print("We found "+str(len(eyes_in_faces)) + " Eyes in this image")
print("Their coordinates and lengths/widths are as follows")
print("=============================")
print(eyes_in_faces)
```

```python
img_with_detections = np.copy(image)

for (x,y,w,h) in eyes_in_faces:
    cv2.rectangle(img_with_detections,(x,y),(x+w,y+h),(255,0,0),5)
    
plt.figure(figsize=(20,10))
plt.imshow(img_with_detections)
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier('../../practice/detector_architectures/haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(gray,1.2,6)
```

```python
print("We found "+str(len(faces)) + " Faces in this image")
print("Their coordinates and lengths/widths are as follows")
print("=============================")
print(faces)
```

```python
img_with_detections = np.copy(image)

for (x,y,w,h) in faces:
    cv2.rectangle(img_with_detections,(x,y),(x+w,y+h),(255,0,0),5)
    
plt.figure(figsize=(20,10))
plt.imshow(img_with_detections)
plt.show()
```
